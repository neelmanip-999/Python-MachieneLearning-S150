✅ 2. Import and Datasets
import seaborn as sns
import matplotlib.pyplot as plt
import: Loads a library into Python.

seaborn: The library you're importing.

as sns: Creates a shortcut name sns for seaborn.

matplotlib.pyplot: Another plotting library. Seaborn is built on top of it.

plt: Alias to refer to pyplot.

tips = sns.load_dataset("tips")
sns.load_dataset(): Loads built-in datasets in seaborn.

"tips": Name of the dataset (a sample restaurant tip dataset).

tips: Variable where the DataFrame is stored.

✅ 3. Bar Plot
sns.barplot(x="sex", y="total_bill", data=tips)
sns.barplot(): Function to create a bar plot.

x="sex": X-axis uses the sex column from the dataset.

y="total_bill": Y-axis uses the total_bill column.

data=tips: Source of the data is the tips DataFrame.

✅ 4. Hist Plot
sns.histplot(tips['total_bill'], bins=10, kde=False)
tips['total_bill']: Selects the total_bill column.

bins=10: Divide the data range into 10 intervals.

kde=False: Disables the KDE curve (default is False).

✅ 5. KDE Plot
sns.kdeplot(tips['total_bill'], shade=True)
kdeplot(): Plots a smooth distribution.

shade=True: Fills the area under the curve.

✅ 6. Scatter Plot
sns.scatterplot(x="total_bill", y="tip", data=tips, hue="sex")
scatterplot(): Creates a scatter plot.

x="total_bill": X-axis values.

y="tip": Y-axis values.

hue="sex": Adds color based on the "sex" column.

✅ 7. Line Plot
sns.lineplot(x="size", y="tip", data=tips)
lineplot(): Connects points with lines.

x, y: Axes values.

✅ 8. Joint Plot
sns.jointplot(x="total_bill", y="tip", data=tips, kind="reg")
jointplot(): Shows scatter + histograms.

kind="reg": Adds a regression line (linear fit).

✅ 9. Pair Plot
sns.pairplot(tips, hue="sex")
pairplot(): Shows all pairwise plots between numeric columns.

hue="sex": Color points by gender.

✅ 10. Box Plot
sns.boxplot(x="day", y="total_bill", data=tips)
boxplot(): Shows quartiles, median, outliers.

✅ 11. Violin Plot
sns.violinplot(x="day", y="total_bill", data=tips)
violinplot(): Box plot + KDE on each side.

✅ 12. Strip Plot
sns.stripplot(x="day", y="total_bill", data=tips, jitter=True)
stripplot(): Plots raw data points.

jitter=True: Slightly moves points so they don’t overlap.

✅ 13. Swarm Plot
sns.swarmplot(x="day", y="total_bill", data=tips)
swarmplot(): Similar to stripplot but avoids overlapping smartly.

✅ 14. Palette (Color Themes)
sns.set_palette("pastel")
set_palette(): Changes the default color theme.

"pastel": A light color palette.

✅ 15. Context
sns.set_context("talk")
set_context(): Controls font size and element scale.

"talk": Increases size for presentations (others: "paper", "notebook", "poster").

✅ 16. Heatmap with Correlation
corr = tips.corr(numeric_only=True)
tips.corr(): Calculates correlation between numeric columns.

numeric_only=True: Ignores non-numeric columns.

sns.heatmap(corr, annot=True, cmap="coolwarm")
heatmap(): Creates a color grid to show correlations.

annot=True: Shows numbers inside squares.

cmap="coolwarm": Blue to red color theme.

✅ 17. Heatmap with Flights Dataset
flights_pivot = flights.pivot(index="month", columns="year", values="passengers")
pivot(): Reshapes the data into a table with month as rows, year as columns.

values="passengers": Fills table values with passengers.
sns.heatmap(flights_pivot, annot=True, fmt="d", cmap="YlGnBu")
fmt="d": Format values as integers.

cmap="YlGnBu": Yellow-Green-Blue color map.

```python
sns.barplot(tips_data, x='day', y='tip', estimator='mean', hue='sex', palette='crest')


# 🔧 sns.barplot(...)
# Seaborn's barplot creates bar plots where the height of each bar represents a summary statistic (mean by default) of a numeric variable.

# 📌 Parameters Explained
# Parameter	Description
# tips_data	The DataFrame containing the data (probably sns.load_dataset('tips')).
# x='sex'	Categorical variable for the x-axis ('Male', 'Female').
# y='tip'	Numerical variable for the y-axis (the tip amount).
# estimator='mean'	Computes the mean tip for each group (default is mean, can be changed to sum, len, np.median, etc.).
# hue='day'	Separates bars by the day (e.g., Thur, Fri, Sat, Sun) using different colors. Each bar group is subdivided by day.
# palette='crest'	Specifies the color palette used for different hue levels. 'crest' is a built-in Seaborn palette with smooth gradients.

# 🔍 What the Plot Shows
# For each sex (Male/Female):

# The plot shows average tip amounts.

# Each bar is further split by day using different colors (legend shown).

# Height of the bar = average tip.

# Color of the bar = day of the week (as per hue).

# Palette = makes the chart visually appealing with a gradient color scheme.

```

```python
sns.barplot(tips_data,y = 'day',x = 'total_bill',estimator='mean',hue='time',palette='viridis',orient='h')
```

```python


sns.histplot(tips_data,y = 'total_bill',kde=True)

# sns.histplot(tips_data,y = 'total_bill',kde=True)

# This line creates a histogram of the total_bill values from the tips_data dataset, plotted vertically, with an optional KDE curve.

# 📌 Parameter Explanation
# Parameter	Description
# tips_data	The Pandas DataFrame containing the data (likely from sns.load_dataset('tips')).
# y='total_bill'	The numerical column to be plotted on the y-axis (vertical histogram).
# kde=True	Adds a KDE curve (Kernel Density Estimate) over the histogram to show the smoothed distribution.
```

```python
sns.histplot(tips_data,x = 'tip',kde=True)

# 📌 Parameter-by-Parameter Explanation
# Parameter	Description
# tips_data	The DataFrame you're using (sns.load_dataset("tips")).
# x='tip'	Plots the distribution of the tip column along the x-axis.
# kde=True	Adds a Kernel Density Estimate (KDE) curve — a smooth version of the histogram that estimates the probability distribution.

# 📊 What the Plot Shows
# The x-axis: range of tip values.

# The y-axis: frequency (number of occurrences) in each bin.

# Histogram bars show how frequently certain tip values appear.

# KDE curve (smooth line) shows the estimated distribution (like a smoothed version of the bars).

```

```python
sns.histplot(tips_data,x = 'tip',y='total_bill', color='green')

# ✅ What This Does
# This creates a 2D histogram (also called a hexbin or
#  heatmap-style histogram) that shows the joint distribution of two numeric variables: tip and total_bill.

# 📌 Parameter-by-Parameter Explanation
# Parameter	Description
# tips_data	The dataset (e.g., sns.load_dataset("tips")).
# x='tip'	Values plotted on the x-axis.
# y='total_bill'	Values plotted on the y-axis.
# _(no kde=True)	This is a 2D histogram, so kde is not applicable.

# 📊 What the Plot Shows
# The plot is divided into 2D bins (like rectangular tiles).

# Each bin contains counts of points that fall into that (x, y) region.

# Color intensity (usually darker = more data points) represents frequency in that bin.

# Think of it as a grid:

# More people who paid around $20 total and tipped around $3? That bin will be darker.
```

```python
sns.scatterplot(data =tips_data,x = 'total_bill',y = 'tip',hue='tip',legend=False)

# ✅ What This Does
# This creates a scatter plot where:

# Each dot represents a data point (i.e., one customer's bill and tip).

# The x-axis shows total_bill, and the y-axis shows tip.

# The color (hue) of each point is determined by the tip amount.

# No legend is shown (legend=False).

# 📌 Parameter-by-Parameter Explanation
# Parameter	Meaning
# data=tips_data	The DataFrame (sns.load_dataset('tips')).
# x='total_bill'	Values for the x-axis (bill amount).
# y='tip'	Values for the y-axis (tip amount).
# hue='tip'	Colors each point based on the tip amount (continuous color gradient).
# legend=False	Removes the legend from the plot (useful if too many unique tip values).

# 🔍 What the Plot Shows
# You get a scatter of points where:

# Points are closer to the bottom-left for small bills and tips.

# Points move up/right for higher bills and tips.

# Color varies by the tip amount (dark/light based on value).

# Color gradient helps see how tip amounts vary across the dataset.
```

```python
sns.jointplot(tips_data,x = 'total_bill',y = 'tip',kind='hex')


# 🔍 What jointplot Does
# jointplot() creates a combined plot with:

# A main plot showing the relationship between two variables.

# Marginal plots (on top and right) showing individual distributions.

# 📌 Parameter-by-Parameter Explanation
# Parameter	Description
# tips_data	Your dataset (sns.load_dataset('tips')).
# x='total_bill'	Plots total bill on the x-axis.
# y='tip'	Plots tip amount on the y-axis.
# kind='hex'	Uses a hexbin plot (2D histogram using hexagonal bins).

# 🧊 What is a Hexbin Plot?
# A hexbin plot divides the plot area into hexagonal bins and counts how many data points fall into each bin.

# Darker hexagons = more data points.

# Great for large datasets where scatter plots look messy.

# ✅ What This Plot Shows
# Center plot: hexbin showing where total_bill and tip amounts cluster.

# Top/Right plots: histograms showing the distributions of total_bill and tip individually.

# Ideal for spotting density patterns and correlations between the two variables.

```

```python
sns.jointplot(tips_data,x = 'total_bill',y = 'tip',kind='scatter',hue='sex')
```

```python
sns.pairplot(tips_data,hue='day')

# 📌 What pairplot Does
# pairplot() creates a matrix of scatter plots (also called a scatterplot matrix) for all numerical columns in the dataset, showing their pairwise relationships.

# Diagonal: histograms or KDE plots (univariate distributions).

# Off-diagonal: scatter plots between variable pairs.

# hue='day': Colors the points by the 'day' column (Thur, Fri, Sat, Sun).

# 🔍 What You Get
# Helps visualize correlation and clusters.

# Shows how total_bill, tip, size, etc. relate to each other.

# Points are colored by day, so you can see differences by day of the week.

```

```python
sns.boxplot(tips_data,x = 'day',y = 'total_bill',hue='sex')

# 📦 What boxplot Does
# A box plot (also called a box-and-whisker plot) summarizes the distribution of a numerical variable across categories — showing:

# Median

# Interquartile Range (IQR): 25th to 75th percentiles

# Whiskers: Range (usually within 1.5×IQR)

# Outliers: Data points outside the whiskers

# 📌 Parameter-by-Parameter Explanation
# Parameter	Description
# tips_data	The DataFrame (sns.load_dataset('tips')).
# x='day'	Categorical variable on x-axis (e.g., Thur, Fri, Sat, Sun).
# y='total_bill'	Numerical variable to plot (total bill amount).
# hue='sex'	Splits each box into two groups (Male, Female) using color.

# 📊 What the Plot Shows
# For each day, you get two box plots side by side:

# One for Male

# One for Female

# You can compare:

# Central tendencies (medians)

# Spread of the data

# Outliers

# Differences in total bills between genders per day
```

```python
sns.violinplot(tips_data,x = 'day',y = 'total_bill',hue='sex')

# 🎻 What violinplot Does
# A violin plot shows the distribution of a numeric variable (total_bill) across categories (day), using a combination of:

# Box plot (showing median, quartiles, etc.)

# KDE (smoothed probability density of the data)

# It’s like a fatter, more informative box plot.

# 📌 Parameter-by-Parameter Explanation
# Parameter	Description
# tips_data	The DataFrame (likely from sns.load_dataset("tips")).
# x='day'	Categorical variable for the x-axis (e.g., Thur, Fri, Sat, Sun).
# y='total_bill'	Numeric variable for the y-axis (total bill amounts).
# hue='sex'	Splits each violin into two by gender (Male/Female) with different colors.

# 📊 What the Plot Shows
# For each day, you get:

# A violin shape showing the distribution of total_bill.

# Inside each violin:

# Split left/right by sex (Male/Female).

# Wider areas = more data density.

# A white dot = median.

# A thick bar = interquartile range (25%–75%).

# A thin line = overall data range (except outliers).
```

```python
sns.violinplot(tips_data,x = 'day',y = 'total_bill',split=True,hue = 'sex')

# ❌ Problem: split=True raises an error unless the hue variable has exactly 2 unique categories
# Also, you must set inner if you want control over the internal representation.

# ✅ Correct Usage (When hue='sex')
# Since tips_data['sex'] has exactly 2 categories: 'Male' and 'Female', split=True is valid.

# But Seaborn requires split=True only when you want to combine both hue categories into a single violin shape per group, split left/right.

# 🔍 Explanation of Parameters
# Parameter	Description
# tips_data	The dataset (sns.load_dataset('tips')).
# x='day'	Days of the week (Thur, Fri, Sat, Sun) on the x-axis.
# y='total_bill'	Numeric variable on the y-axis.
# hue='sex'	Splits data into Male vs Female.
# split=True	Combines both sexes into one violin per day, split left/right. Requires hue to have exactly 2 categories.


```

```python
sns.stripplot(tips_data,x = 'day',y = 'total_bill')

# 📌 What stripplot Does
# stripplot() creates a scatter plot of individual data points for categorical variables — it's great for showing the distribution and spread of the data.

# ✅ Parameter Breakdown
# Parameter	Description
# tips_data	Your dataset (sns.load_dataset("tips")).
# x='day'	Categorical variable on the x-axis (Thur, Fri, Sat, Sun).
# y='total_bill'	Numeric variable on the y-axis (the amount of the total bill).

# 📊 What This Plot Shows
# Each dot represents one observation (i.e., one customer's bill).

# For each day, dots are plotted at the corresponding total_bill value.

# Dots may overlap if many points have similar values.

# This gives a sense of data concentration — how many customers had bills around a certain value.

```

```python
sns.swarmplot(tips_data,x = 'day',y = 'total_bill')

# 🐝 What swarmplot Does
# A swarmplot() is similar to stripplot(), but it avoids overlapping points by smartly arranging them so each dot is visible.

# It gives a clear view of the distribution of individual data points while preserving the data density.

# 📌 Parameter-by-Parameter Explanation
# Parameter	Description
# tips_data	Your DataFrame (e.g., from sns.load_dataset('tips')).
# x='day'	Categorical variable on the x-axis (Thur, Fri, Sat, Sun).
# y='total_bill'	Numerical variable on the y-axis (total bill values).

# 📊 What the Plot Shows
# Each point = one row in the dataset.

# Points are spaced out automatically to avoid overlap.

# Unlike stripplot(), which might stack dots on top of each other, swarmplot() spreads them out horizontally while maintaining their actual y-value.
```

```python
sns.heatmap(tips_data.corr(numeric_only = True),annot = True)

# ✅ What This Code Does
# This creates a correlation heatmap of all the numerical columns in the tips_data dataset.

# 📌 Parameter-by-Parameter Breakdown
# 🔹 tips_data.corr(numeric_only=True)
# Computes the correlation matrix of numerical columns only.

# Correlation values range from -1 to 1:

# +1 → perfect positive correlation

# 0 → no correlation

# -1 → perfect negative correlation

# Typical numerical columns in tips_data include:

# total_bill

# tip

# size

# 🔹 sns.heatmap(..., annot=True)
# Draws a heatmap from the correlation matrix.

# annot=True: Shows the correlation values inside each cell.

# ✅ Sample Output (Conceptually)
# total_bill	tip	size
# total_bill	1.00	0.68	0.60
# tip	0.68	1.00	0.49
# size	0.60	0.49	1.00

# Darker/lighter colors show the strength of correlation.
```
