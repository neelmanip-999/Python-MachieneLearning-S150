âœ… figure and subplot â€“ For Multiple Plots
ðŸ”¹ figure: Creates a new figure window.
ðŸ”¹ subplot: Divides the figure into a grid and places plots in each section.

import matplotlib.pyplot as plt
import numpy as np

x = np.arange(1, 6)
y1 = x ** 2
y2 = x ** 3

plt.figure(figsize=(8, 4))

plt.subplot(1, 2, 1)  # (rows, columns, position)
plt.plot(x, y1)
plt.title("Square")

plt.subplot(1, 2, 2)
plt.plot(x, y2)
plt.title("Cube")

plt.tight_layout()
plt.show()

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
X = np.linspace(-2,2,1000)
ySin= np.sin(2*np.pi*X)
yCos= np.cos(2*np.pi*X)

ytan= np.tan(2*np.pi*X)
ycot= yCos/ySin

fig,axes = plt.subplots(2,2,figsize=(12,6))
#fig,((ax1,ax2),(ax3,ax4)) = plt.subplots(2,2,figsize=(12,6))
axes[0,0].plot(X,ySin)
axes[0,0].grid()
axes[0,0].set_title('Sin wave')

axes[0,1].plot(X,yCos)
axes[0,1].grid()
axes[0,1].set_title("Cos Wave")

axes[1,0].plot(X,ytan)
axes[1,0].grid()
axes[1,0].set_ylim((-5,5))
axes[1,0].set_title("Tan Wave")

axes[1,1].plot(X,ycot)
axes[1,1].grid()
axes[1,1].set_ylim((-5,5))
axes[1,1].set_title("Cot Wave")

plt.suptitle("Tignometric Functions")
plt.subplots_adjust(hspace=0.3)
plt.show()





```
