Gradient Descent is an optimization algorithm used in machine learning and deep learning to minimize the loss function (error) and improve the model‚Äôs predictions.


üîç What Is It?
Gradient Descent is like climbing down a hill to find the lowest point (minimum value). In machine learning, that hill is the loss function, and our goal is to find the best model parameters (like weights) that minimize this loss.


üìâ How It Works (Step-by-Step)
Initialize parameters randomly.

Compute the loss (e.g., MSE, Cross-Entropy).

Calculate gradient (partial derivatives w.r.t. each parameter).

Update parameters using the gradient.

Repeat until convergence (loss stops decreasing or max iterations reached).

‚ö†Ô∏è Common Challenges
Learning rate 
ùõº
Œ± too high: overshooting the minimum.

Learning rate too low: slow convergence.

Local minima or saddle points can trap the algorithm.

Vanishing gradients (in deep networks).






```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def fn(X):
    return 3*X**2-4*X+5

def derFn(X):
    return 6*X-4
```

```python
X = np.linspace(-2.5,4)
y = fn(X)
```

```python
plt.plot(X,y)
```

```python
xm = 3
alpha = 0.25
xml = [xm]
```

```python
for i in range(20):
    xm = xm - alpha*derFn(xm)
    xml.append(xm)
```

```python
plt.plot(X,y)
plt.plot(xml,fn(np.array(xml)),'r')
plt.ylim(0,40)
plt.xlim(-3,5)
plt.show()
```
